## 17. 
Build a dataset with every possible (UserID, ProductID) pair (cross join)

```python
users
```

```python
products
```

```python
#since there are no key commont to them, lets create one
users1 = users.copy()
users1['key']= 0
products1 = products.copy()
products1['key']=0
user_product_merged = users1.merge(products1,how='outer',on='key')[['UserID','ProductID']]
user_product_merged
```

## 18. 
Determine how much quantity of each product was purchased by each user

```python
qty = transactions.groupby(['UserID','ProductID'])['Quantity'].agg('sum').reset_index()
userwise_prod = user_product_merged.merge(qty,how='left',on=('UserID','ProductID'))
userwise_prod.fillna(0)
```
